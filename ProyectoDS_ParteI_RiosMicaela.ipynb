```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

##  Archivo CSV en Gitgub
df = pd.read_csv('https://raw.githubusercontent.com/MicaelaER/CODERHOUSE-FundamentosData/main/heart_attack_dataset.csv')

# Info acerca de los datos
df.info()
df.describe()
df.head()


# Crear grficos con Matplotlib
plt.figure(figsize=(12, 8))

# Grfico 1: Distribucin de la presin arterial por gnero
plt.subplot(2, 2, 1)
plt.hist([df[df['Gender'] == 'Male']['Blood Pressure (mmHg)'], df[df['Gender'] == 'Female']['Blood Pressure (mmHg)']], bins=10, label=['Male', 'Female'], edgecolor='black')
plt.title('Distribucin de la Presin Arterial por Gnero')
plt.xlabel('Presin Arterial (mmHg)')
plt.ylabel('Frecuencia')
plt.legend()

# Grfico 2: Colesterol por tipo de dolor de pecho
plt.subplot(2, 2, 2)
sns.boxplot(data=df, x='Chest Pain Type', y='Cholesterol (mg/dL)')
plt.title('Colesterol por Tipo de Dolor de Pecho')
plt.xlabel('Tipo de Dolor de Pecho')
plt.ylabel('Colesterol (mg/dL)')

# Grfico 3: Edad por estado de diabetes
plt.subplot(2, 2, 3)
sns.barplot(data=df, x='Has Diabetes', y='Age')
plt.title('Edad por Estado de Diabetes')
plt.xlabel('Tiene Diabetes')
plt.ylabel('Edad')

plt.tight_layout()
plt.show()

# Crear grficos con Seaborn
plt.figure(figsize=(12, 8))

# Grfico 4: Presin arterial por estado de fumador
plt.subplot(2, 2, 1)
sns.violinplot(data=df, x='Smoking Status', y='Blood Pressure (mmHg)')
plt.title('Presin Arterial por Estado de Fumador')
plt.xlabel('Estado de Fumador')
plt.ylabel('Presin Arterial (mmHg)')

# Grfico 5: Colesterol por tratamiento
plt.subplot(2, 2, 2)
sns.barplot(data=df, x='Treatment', y='Cholesterol (mg/dL)')
plt.title('Colesterol por Tratamiento')
plt.xlabel('Tratamiento')
plt.ylabel('Colesterol (mg/dL)')
plt.xticks(rotation=90)

# Grfico 6: Edad por gnero
plt.subplot(2, 2, 3)
sns.boxplot(data=df, x='Gender', y='Age')
plt.title('Edad por Gnero')
plt.xlabel('Gnero')
plt.ylabel('Edad')

plt.tight_layout()
plt.show()
```
